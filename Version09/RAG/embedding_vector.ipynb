```python
# 匯入套件
import akasha
import akasha.helper as ah

# 設定模型資訊
emb_obj = ah.handle_embeddings("openai:text-embedding-ada-002", verbose=True)
```

```python
# 將文字轉成向量
hello_vector = emb_obj.embed_query("哈囉，世界！")

display(hello_vector)
display('這個向量的長度是：' + str(len(hello_vector)))
```

```python
# 將文字轉成向量
poetry_vector = emb_obj.embed_query("""
白日依山盡，黃河入海流，
欲窮千里目，更上一層劉。""")

display(poetry_vector)
display('這個向量的長度是：' + str(len(poetry_vector)))
```

```python
import numpy as np

# 打招呼
poetry_vector = emb_obj.embed_query("""
白日依山盡，黃河入海流，欲窮千里目，更上一層劉。""")

# 明星型帥哥
handsome_1_vector = emb_obj.embed_query("彭于晏長得很帥")
handsome_2_vector = emb_obj.embed_query("金城武是帥哥")

# 禮貌型帥哥
fake_handsome_vector = emb_obj.embed_query("早餐店老闆娘都叫我帥哥")

# 兩兩比較差別
print('「一首詩」與「禮貌型帥哥」的距離: ' + str(np.mean(np.abs(np.array(poetry_vector) - np.array(fake_handsome_vector)))))   
print('「禮貌型帥哥」與「明星型帥哥-1」的距離: ' + str(np.mean(np.abs(np.array(fake_handsome_vector) - np.array(handsome_1_vector)))))
print('「禮貌型帥哥」與「明星型帥哥-2」的距離: ' + str(np.mean(np.abs(np.array(fake_handsome_vector) - np.array(handsome_2_vector)))))
print('「明星型帥哥-1」與「明星型帥哥-2」的距離: ' + str(np.mean(np.abs(np.array(handsome_1_vector) - np.array(handsome_2_vector)))))
```
